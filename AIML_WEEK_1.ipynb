```python
import matplotlib.pyplot as plt

people = ['radha','arjun','krish','sandy','nita']
age = [24,28,46,54,32]
weight = [50,60,70,40,56]
height = [170,175,180,150,175]

plt.scatter(weight,height)
plt.title("relationship between weight and height of patients")
plt.ylabel("height in meters")
plt.xlabel("weight of patients in kgs")
plt.show()

plt.scatter(age,weight)
plt.title("relationship between age and weight of patients")
plt.ylabel("weight in meters")
plt.xlabel("age of patients in years")
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.title("people's weight in kgs",fontsize=30,fontstyle='italic')

plt.bar(x=people,height=weight,width=0.6,color='red',edgecolor='k',alpha=0.6)

plt.xlabel("people",fontsize=15)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))

plt.hist(weight,color='red',edgecolor = 'k',alpha=0.75)

plt.title("histogram of patients weight",fontsize=19)
plt.xlabel("weight in kgs",fontsize=15)
plt.show()
```
